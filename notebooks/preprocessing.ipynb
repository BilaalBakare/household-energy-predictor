```python
import pandas as pd
import numpy as np

DATA_RAW = '../data/household_power_consumption_eda.csv'
df = pd.read_csv(DATA_RAW, sep=',', low_memory=False)

pd.set_option('display.precision', 2)  
```

### Remove corrupt rows

```python
df.dropna(subset=['Sub_metering_3'], inplace=True)
```

```python
df[df['Sub_metering_3'].isnull()].head()
```

### Merging the date and time columns

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True)
df.drop(columns=['Date', 'Time'], inplace=True)
```

```python
df.head()
```

### Decomposing the datetime column

```python
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Day'] = df['Datetime'].dt.day
df['Hour'] = df['Datetime'].dt.hour
df['Minute'] = df['Datetime'].dt.minute    

df.drop(columns=['Datetime'], inplace=True)
```

```python
df.head()
```

```python
df.to_csv('../data/household_power_consumption_preprocess_01.csv', index=False)
```

Benchmark training results:

RMSE: 0.0305
R2 percentage: 99.91757647499917
Mean absolute percentage error: 0.020562466632309068

### dropping the Global_intensity column

Since it is noticed that the Global_intensity column has a '1' correlation with the target column, It is decided to drop the column

```python
df.drop(columns=['Global_intensity'], inplace=True)
```
